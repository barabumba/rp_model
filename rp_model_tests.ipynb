# Сравнение ожидаемой  и фактической дисперсии

Ожидаемая дисперсия определяется как $$\sigma^{2}=\frac{\Omega_{0}}{2\pi}\frac{N_{0}}{2}\int\limits_{-\infty}^{\infty}f(x)dx,$$
что после дискретизации приведёт к выражению
$$\sigma^{2}=\sigma_{0}^{2} \frac{2M_{0}}{N}\int\limits_{0}^{N/2M_{0}}f(x)dx=\sigma_{0}^{2} \frac{M_{0}}{N},$$
где $\sigma_{0}^{2}=1$ - дисперсия формирующего шума
